```python
import pandas as pd
import datetime
import numpy as np
```

```python
d1 = pd.read_csv('ecommerce/ShopSpectra Transaction Dataset.csv', index_col=0)
d1
```

```python
d2 = pd.read_csv('ecommerce/ShopSpectra Transaction Dataset 2.csv', index_col=0)
d2
```

```python
d3 = pd.read_csv('ecommerce/ShopSpectra Transaction Dataset 3.csv', index_col=0)
d3
```

```python
df = pd.concat([d1, d2, d3], ignore_index=True)
```

```python
# aggregated spending by userID 

df_features = df.groupby('UserID').agg({
    'TransactionAmount': ['sum', 'mean', 'max', 'min', 'std'],
    'UserID': 'count'
}).reset_index()
df_features.columns = ['UserID', 'TotalAmount', 'AvgAmount', 'MaxAmount', 'MinAmount', 'AmountStd', 'TransactionCount']
df_features
```

```python
# Force conversion of TransactionDate to datetime, coerce errors to NaT
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], errors='coerce')

# Filter out rows where TransactionDate could not be converted
df = df.dropna(subset=['TransactionDate'])

# Calculate Recency properly - this is the last time they made a transaction
latest_date = df['TransactionDate'].max()  # Ensure latest_date is datetime
df['Recency'] = (latest_date - df['TransactionDate']).dt.days
```

```python
df
```

```python
# took each customer and determined the time they preffered to shop
df_time = df.groupby('UserID').agg(
    Recency=('Recency', 'min'),
    PreferredHour=('TransactionHour', lambda x: x.mode()[0])
).reset_index()
df_time
```

```python
# took each customer payment type and method to pay
df_behavior = df.groupby('UserID').agg(
    FraudRate=('Is_Fraud', 'mean'),
    DeclineRate=('Is_Declined', 'mean'),
    UniquePaymentMethods=('PaymentMethod', pd.Series.nunique),
     PreferredPayment=('PaymentMethod', lambda x: x.mode()[0]),
    PreferredDevice=('DeviceType', lambda x: x.mode()[0])
).reset_index()
```

```python
df_behavior
```

```python
# took each customer location of most frequency of transactions
df_location = df.groupby('UserID').agg(
Lat=('Latitude', lambda x: x.mode()[0]),
Long=('Longitude', lambda x: x.mode()[0]),
).reset_index()
df_location
```

```python
# customer demos and the average amount per item purchased
df_demographics = df.groupby('UserID').agg(
    AvgCustomerAge=('CustomerAge', 'mean'),
    AvgAccountAge=('AccountAgeDays', 'mean'),
    AvgBasketSize=('TransactionAmount', lambda x: np.sum(x) / np.sum(df.loc[x.index, 'Quantity']))
).reset_index()
```

```python
df_demographics
```

```python
# Merchant Category Features
df_merchant = df.groupby(['UserID', 'MerchantCategory']).size().unstack(fill_value=0).reset_index()
df_merchant
```

```python
# Transaction Status Features
df_status = df.groupby(['UserID', 'TransactionStatus']).size().unstack(fill_value=0).reset_index()
df_status['CompletedRate'] = df_status['Completed'] / df_status.iloc[:, 1:].sum(axis=1)
df_status['FailedRate'] = df_status['Failed'] / df_status.iloc[:, 1:].sum(axis=1)
df_status['PendingRate'] = df_status['Pending'] / df_status.iloc[:, 1:].sum(axis=1)
df_status
```

```python
# merging all the features together to aggregate at the customer level

df_features = df_features.merge(df_time, on='UserID', how='left')
df_features = df_features.merge(df_behavior, on='UserID', how='left')
df_features = df_features.merge(df_location, on='UserID', how='left')
df_features = df_features.merge(df_demographics, on='UserID', how='left')
df_features = df_features.merge(df_merchant, on='UserID', how='left')
df_features = df_features.merge(df_status[['UserID', 'CompletedRate', 'FailedRate', 'PendingRate']], on='UserID', how='left')
```
